# **Desafio**


Fomos contratados por um grande banco para desenvolver o seu novo sistema. Esse banco deseja modernizar suas operações e para isso escolheu a linguagem Python. Para a primeira versão do sistema devemos implementar apenas 3 operações:  depósito, saque e extrato.

## Operação de depósito

Deve ser possível depositar valores positivos para a minha conta bancária. A v1 do projeto trabalha apenas com 1 usuário, dessa forma não precisamos nos preocupar em identificar qual é o número da agência e conta bancária. Todos os depósitos devem ser armazenados em uma variável e exibidos na operação de extrato.

## Operação de Saque

O sistema deve permitir realizar 3 saques diários com limite máximo de R$ 500,00 por saque. Caso o usuário não tenha saldo em conta, o sistema deve exibir uma mensagem informando que não será possível sacar o dinheiro por falta de saldo. Todos os saques devem ser armazenados em uma variável e exibidos na operação de extrato.

## Operação de extrato

Essa operação deve lista todos os depósitos e saques realizados na conta. No fim da listagem deve ser exibido o saldo atual da conta.

Os valores devem ser exibidos utilizando o formato R$ xxx.xx,

exemplo:

1500.45 = R$ 1500.45


```python
print(" Sistema Bancário ".center(60, "="))

NUM_DEPOSITOS: int = 0
NUM_SAQUES: int = 0
saque:float = 0
saldo: float = 0
deposito: float = 0
SAQUE_MAXIMO: float = 500
SAQUES_DIARIOS = 0
historico:str = ""

while True:
  print("""
  [1] - Sacar
  [2] - Depositar
  [3] - Visualizar Extrato
  [4] - Sair
  """)

  opcao = input("->").strip()
  match opcao:
    case "1":
      if saldo == 0:
        print("Você não tem Saldo ainda para sacar!")
      else:
        if SAQUES_DIARIOS == 3:
          print("Você já atingiu o limete de 3 saques.")
          SAQUES_DIARIOS = 0
        else:
          SAQUES_DIARIOS += 1
          if saque >= 500:
            print("IF - 3")
            print("Infelizmente você não pode sacar mais que 500 em uma transação.")
          else:
            saque = float(input("Informe o valor do saque: "))
            if saque > saldo:
              print("Saque Negado - Você não tem saldo suficiente!")
              historico += ", " + f"Saque: {NUM_SAQUES} - Valor: R$ {saque} - Negado"
            else:
              saldo -= saque
              historico += ", " + f"Saque: {NUM_SAQUES} - Valor: R$ {saque} - Aprovado"
              print("Saque aprovado!")
            NUM_SAQUES += 1
    case "2":
      deposito = float(input("Informe o valor do deposito: "))
      if deposito >= 0:
        saldo += deposito
        print("Deposito realizado com sucesso!")
        if len(historico) == 0:
          NUM_DEPOSITOS += 1
          historico += f"Deposito: {NUM_DEPOSITOS} - Valor: R$ " + str(deposito)
        else:
          historico += ", " + f"Deposito: {NUM_DEPOSITOS} - Valor: R$ " + str(deposito)
        NUM_DEPOSITOS += 1
      else:
        print("O valor do deposito não pode ser negativo!")
    case "3":
      extrato = historico.split(",")
      for x in extrato:
        print(x.strip())
    case "4":
      print("Sair")
      break
```

```python
print(" Sistema Bancário ".center(60, "="))
while True:
  print("""
  [1] - Sacar
  [2] - Depositar
  [3] - Visualizar Extrato
  [4] - Sair
  """)

  opcao = input("->").strip()
  opcao = int(opcao)

  if opcao == 1:
    print("Sacar")
  elif opcao == 2:
    print("Depositar")
  elif opcao == 3:
    print("Visualizar Extrato")
  else:
    print("Sair")
    break
```
