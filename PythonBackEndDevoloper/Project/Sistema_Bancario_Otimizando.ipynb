# Desafio: Criando um sistema bancário

## Objetivo Geral

Separar as funções existentes de saque, depósito e extrato em funções. Criar duas novas funções: cadastrar usuário (cliente) e cadastrar conta bancária.

## Desafio

Precisamos deixar nosso código mais modullarizado, para isso vamos criar funções para as operações existentes: sacar, depositar e visualizar histório. Além disso, para a versão 2 do nosso sistema precisamos criar duas novas funções: criar usuário (cliente do banco) e criar conta corrente (vincular com usuário).

## Separação em funções

Devemos criar funções para todas as operação do sistema. Para exercitar tudo o que aprendemos neste módulo, cada função vai ter uma regra na passagem de argumentos. O retorno e a forma como serão chamadas, pode ser definida por você da forma que achar melhor.

## Saque

A função saque deve receber os argumentos apenas por nome (Keyword only). Sugestaõ de argumentos: saldo, valor, extrato, limite, numero_saques, limite_saques. Sugestão de retorno: saldo e extrato.

## Deposito

A função depósito deve receber os argumentos apenas por posição (positional only). Sugestão de argumentos: saldo, valor, extrato. Sugestão de retorno: saldo e extrato.

## Extrato

A função extrato deve receber os argumentos por posição e nome (positional only e keyword only). Argumentos posicionais: saldo, argumentos nomeados: extrato.

## Novas funções

Precisamos criar duas novas funções: criar usuário e criar conta corrente. Fique a vontade para adicionar mais funções, exemplo: listar contas.

## Criar usuário (cliente)

O programa deve armazenar os usuário em uma lista, um usuário é composto por: nome, data de nascimento, cpf e endereço. O endereço é uma string com o formato: logradouro, nro - bairro - cidade/sigla estado. Deve ser armazenado somente os números do CPF. Não podemos cadastrar 2 usuários com o mesmo CPF.

## Criar conta corrente

O programa deve armazenar contas em uma lista, uma conta é composta por: agência, número da conta e usuario. O número da conta é sequencial, iniciando em 1. O número da agência é fixo: "0001". O usuário pode ter mais de uma, mas uma conta pertence a somente um usuário.

## Dica

Para vincular um usuário a uma conta, filtre a lista de usuários buscando o número do CPF informado para cada usuáiro da lista.x


```python
CONTA = 0

USUARIOS = {
    "nomes": [],
    "dataNascimento": [],
    "cpf": [],
    "enderecos": [],
    "contas": [],
    "agencia": [],
    "tipo_cliente": [],
    "senhas": [],
    "Logado": [],
    "historico": [],
    "temCartao": [],
    "saque": [],
    "saldo": []
}

def login(cpf: str, senha: str) -> str:
  if cpf in USUARIOS['cpf']:
    pos = USUARIOS['cpf'].index(cpf)
    print(pos)
    if senha == USUARIOS['senhas'][pos]:
      USUARIOS["Logado"][pos] = True
      return USUARIOS["Logado"], pos
    else:
      return False, -1
  else:
    return False, -1

def criarContaCorrente(usuario, cpf):
  if cpf in USUARIOS['cpf']:
    numeroConta = len(USUARIOS['contas']) + 1
    USUARIOS["contas"].append([cpf, usuario, numeroConta, "0001"])
    return f"Conta Criada Com Sucesso!"
  else:
    return f"Usuario precisa ser cadastrado antes de criar a sua conta corrente."

def endereco(logradouro, num, bairro, cidade, sigla_estado) -> str:
  return f"{logradouro}, {num} - {bairro} - {cidade}/{sigla_estado}"

def dataNascimento(dia, mes, ano) -> str:
  return f"{dia}/{mes}/{ano}"

def criarUsuario(nome, dia, mes, ano, cpf, logradouro, num, bairro, cidade, sigla_estado, senha, tipoCliente) -> None:
  global CONTA
  CONTA = CONTA + 1
  if cpf not in USUARIOS['cpf']:

    USUARIOS['nomes'].append(nome)
    USUARIOS['dataNascimento'].append(dataNascimento(dia, mes, ano))
    USUARIOS['cpf'].append(cpf)
    USUARIOS['enderecos'].append(endereco(logradouro=logradouro, num=num, bairro=bairro, cidade=cidade, sigla_estado=sigla_estado))
    USUARIOS['senhas'].append(senha)
    USUARIOS['temCartao'].append(False)

    USUARIOS['Logado'].append(False)
    USUARIOS['contas'].append(CONTA)
    USUARIOS['saque'].append(0)
    USUARIOS['deposito'].append(0)
    USUARIOS['historico'].append("")

  else:
    print(f"Usuário {nome} já foi cadastrado!")

def saque():
  cpf = input('Informe o seu CPF: ').strip()
  quantia = float(input("Informe o valor do saque: "))
  pos = USUARIOS['cpf'].index(cpf)
  saldo = USUARIOS['saldo'][pos]
  while quantia > saldo:
    print("Você não tem saldo para saque.")
  else:
    USUARIOS['saldo'][pos] -= quantia
    print("Saque realizado com sucesso!")
    USUARIOS['historico'][pos] += f"Saque realizado no valor de {quantia} "

def deposito():
  cpf = input("Informe o seu CPF: ").strip()
  pos = USUARIOS["cpf"].index(cpf)
  while True:
    quantia = float(input("Informe o valor do seu deposito: "))
    if quantia > 0:
      USUARIOS['saldo'][pos] += quantia
      USUARIOS['historico'][pos] += f"Deposito realizado com sucesso no valor de {quantia} "
      break
    else:
      print("Informe um valor positivo para o deposito.")

def extrato():
  cpf = input("Informe o seu cpf: ").strip()
  pos = USUARIOS["cpf"].index(cpf)
  print(f"O historico do usuario {USUARIOS['nomes'][pos]} é {USUARIOS['historico'][pos]}")

def trocarSenha(usuario: str, cpf: str, isLogado: bool, senhaAntiga: str, novaSenha: str, repitirNovaSenha: str) -> str:
  if isLogado:
    for chaves, valor in USUARIOS.items():
      if cpf in valor:
        v = valor.index(cpf)
        if senhaAntiga == USUARIOS[v]:
          if novaSenha == repitirNovaSenha:
            USUARIOS[v] = novaSenha
            return f"senha de {usuario} redefinidas com sucesso!"
          else:
            return f"Nova senha diferente da Senha repetida!"
        else:
          return f"Senha antiga errada!"
      else:
        return f"{cpf} incorreto!"

def esqueceuSenha(cpf) -> None:
  pos = USUARIOS['cpf'].index(cpf)
  while True:
    senhaAntiga = USUARIOS["senhas"][pos]
    novaSenha = input("Informe a nova senha: ").strip()
    repitaNovaSenha = input("Informe a nova senha: ").strip()
    if novaSenha == repitaNovaSenha and novaSenha != senhaAntiga:
      print("Senha alterada com sucesso!")
      break
    else:
      if novaSenha != repitaNovaSenha:
        print("Você digitou senhas diferentes!")
      elif novaSenha == senhaAntiga:
        print("Sua senha nova não pode ser a mesma de antes.")

def alterarCadastro(cpf) -> None:
  pos = USUARIOS['cpf'].index(cpf)
  print("""Informe o cadastro vc quer alterar:\n[01] - Nome\n[02] - Data de Nascimento\n[03] - CPF\n[04] - Endereco""")
  opcao = int(input("Escolha sua opção: "))
  if opcao == 1:
    USUARIOS["nomes"][pos] = input("Digite o seu nome: ").strip().title()
  elif opcao == 2:
    dia = input("Informe o dia: ").strip().title()
    mes = input("Informe o mes: ").strip().title()
    ano = input("Informe o ano: ").strip().title()
    USUARIOS["dataNascimento"][pos] = dataNascimento(dia, mes, ano)
  elif opcao == 3:
    USUARIOS["cpf"][pos] = input("Informe o novo CPF: ").strip()
  elif opcao == 4:
    logradouro = input("Informe o lougradoro: ").strip().title()
    numCasa = input("Informe o número da casa: ").strip()
    bairro = input("Informe o nome do bairro: ").strip().title()
    cidade = input("Informe o nome da cidade: ").strip().title()
    sigla_estado = input("Informe a sigla do estado: ").strip().title()

    USUARIOS["endereco"][pos] = endereco(logradouro, numCasa, bairro, cidade, sigla_estado)

def pediCartao(cpf) -> None:
  if cpf in USUARIOS["cpf"]:
    pos = USUARIOS['cpf'].index(cpf)
    if USUARIOS["Logado"]:
      USUARIOS["temCartao"][pos] = True
      print("Seu cartão foi enviado.")
  else:
    print("Esse usuário não está cadastrado!")

def logoff() -> None:
  cpf = input("Informe o seu CPF: ").strip()
  pos = USUARIOS['cpf'].index(cpf)
  if USUARIOS['Logado'][pos] == True:
    USUARIOS["Logado"] = False
    print(f'Usuário {USUARIOS["nomes"][pos]} saiu.')
  else:
    print(f"O usuário {USUARIOS['nomes'][pos]} não está logado.")

while True:
  print("""[00] - Login
[01] - Criar Conta
[02] - Criar usuario
[03] - Saque
[04] - Deposito
[05] - Extrato
[06] - Trocar Senha
[07] - Esqueceu a Senha
[08] - Alterar cadastro
[09] - Pedir Cartão
[10] - Fim da operação
""")
  opcao = int(input("Informe seu a transação: ").strip())
  match opcao:
    case 0:
      cpf = input("Informe o CPF: ").strip()
      senha = input("Informe a SENHA: ").strip()
      status, pos = login(cpf, senha)
      if status == False and pos == -1:
        print("Login Não Efetuado")
      else:
        print("OK")
    case 1:
      nome = input("Informe nome do usuario: ").strip().title()
      cpf = input("Informe o CPF do Usuario: ").strip()

      print(criarContaCorrente(nome, cpf))
    case 2:
      print("Vamos criar um novo usuario!")
      nome = input("Informe o nome completo: ").strip().title()

      print("Data de nascimento")
      dia = input("O dia de nascimento: ").strip()
      mes = input("O mês de nascimento: ").strip()
      ano = input("O ano de nascimento: ").strip()

      cpf = input("Informe o seu CPF: ").strip()

      print("Informe seus dados residêncial")
      logradouro = input("Informe seu logradouro: ").strip().title()
      num = input("Informe seu número: ").strip()
      bairro = input("O nome do bairro: ").strip().title()
      cidade = input("O nome da cidade: ").strip().title()
      sigla_estado = input("Sigla do Estado: ").strip().title()

      senha = input("Agora informe uma senha: ").strip()
      tipoCliente = input("Informe o tipo de cliente: ").strip()

      criarUsuario(nome, dia, mes, ano, cpf, logradouro, num, bairro, cidade, sigla_estado, senha, tipoCliente)

    case 3:
      saque()
    case 4:
      deposito()
    case 5:
      extrato()
    case 6:
      trocarSenha()
    case 7:
      esqueceuSenha()
    case 8:
      alterarCadastro()
    case 9:
      pediCartao()
    case 10:
      logoff()
      break
```
